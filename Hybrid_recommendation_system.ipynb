```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 1: Create the TF-IDF matrix using the 'highlights' column
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(product_info1['highlights'])
# Displaying the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Step 2: Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Create a DataFrame for easy viewing of cosine similarity
cosine_sim_df = pd.DataFrame(cosine_sim, index=product_info1['product_name'], columns=product_info1['product_name'])
```

#### Removing Outliers

```python
mean_similarities = cosine_sim_df.mean(axis=1)

# Create a DataFrame with products and their mean similarity scores
similarity_summary = pd.DataFrame({
    'product_name': mean_similarities.index,
    'mean_similarity': mean_similarities.values
})

# Sort by mean similarity to find potential outliers
similarity_summary = similarity_summary.sort_values(by='mean_similarity')
```

```python
# Calculate mean and standard deviation
mean_similarity = similarity_summary['mean_similarity'].mean()
std_dev_similarity = similarity_summary['mean_similarity'].std()

# Define a threshold for outliers (mean - 1 standard deviation)
threshold = mean_similarity - std_dev_similarity

# Identify outliers
outliers = similarity_summary[similarity_summary['mean_similarity'] < threshold]
print(outliers)
```

```python
outlier_products = product_info1[product_info1['product_name'].isin(outliers['product_name'])]
pd.set_option('display.max_colwidth', None)  # Allows full width for each column
outlier_products[['product_name', 'highlights']]
```

```python
# Defining content recommendation function 
def recommend_products(product_name, cosine_sim_df, product_info_df, n=5):
    """
    Recommend similar products based on cosine similarity.
    
    Parameters:
    - product_name (str): The name of the product for which recommendations are sought.
    - cosine_sim_df (pd.DataFrame): DataFrame containing the cosine similarity matrix.
    - product_info_df (pd.DataFrame): DataFrame containing product information (like highlights).
    - n (int): Number of recommendations to return.

    Returns:
    - List of recommended product names.
    """
    # Check if the product is in the DataFrame
    if product_name not in cosine_sim_df.index:
        return f"Product '{product_name}' not found in the database."

    # Get the index of the product
    product_idx = cosine_sim_df.index.get_loc(product_name)

    # Get the similarity scores for this product
    similarity_scores = cosine_sim_df.iloc[product_idx]

    # Sort the scores in descending order and get the top N
    similar_indices = similarity_scores.argsort()[::-1][1:n + 1]  # Skip the first one (itself)

    # Get the product names of the top N similar products
    similar_products = cosine_sim_df.index[similar_indices].tolist()

    return similar_products
```

```python
# Fuction for recommendation
def get_recommendations(user_id, pred_df, n=10):
    # Ensure the user_id is in the correct format (string, in this case)
    user_id = str(user_id)
    
    # Keeping predictions only for the given user_id
    pred_subset = pred_df[pred_df['uid'] == user_id]

     # Check how many predictions exist for this user
    if len(pred_subset) == 0:
        print(f"No predictions available for user {user_id}")
        return []
    
    # Sorting based on predicted rating (est) in descending order
    pred_subset = pred_subset.sort_values(by='est', ascending=False)
    
    # Selecting the top 'n' products
    top_product_ids = pred_subset.head(n)['iid'].to_list()
    
    # Getting product names for the top recommendations
    top_recommendations = product_info2[product_info2['product_id'].isin(top_product_ids)]['product_name'].to_list()
    
    # If not enough recommendations, print a message
    if len(top_recommendations) < n:
        print(f"Only {len(top_recommendations)} products available for user {user_id}.")
    
    return top_recommendations
```

### Hybrid Recommendation

```python
product_info_copy = product_info1.copy()
product_info_copy.set_index('product_name', inplace=True)
```

```python
def recommend_products(product_name, cosine_sim_df, product_info_df, n=5):
    # Check if the product_name exists in product_info_df
    if product_name not in product_info_df.index:
        return [], []

    # Retrieve the similarity scores
    similarity_scores = cosine_sim_df.loc[product_name]

    # Ensure similarity_scores is a Series
    if isinstance(similarity_scores, pd.DataFrame):
        similarity_scores = similarity_scores.squeeze()  # Convert to Series if it's a DataFrame

    # Check for zero similarity scores, avoiding ambiguous truth value error
    if similarity_scores.isnull().all() or (similarity_scores == 0).all():
        print(f"No similar products found for {product_name} based on similarity scores.")
        return [], []

    # Sort the similarity scores and get the top N similar products
    similar_indices = similarity_scores.argsort()[-(n + 1):-1][::-1]  # Top N products, excluding the product itself
    similar_products = product_info_df.iloc[similar_indices].index.tolist()  # Convert to list

    return similar_products, similarity_scores.iloc[similar_indices]

def get_collaborative_score(user_id, pred_df, n=5):
    # Check if user_id exists in the 'uid' column of pred_df
    if user_id not in pred_df['uid'].values:
        return None, f"No ratings found for user '{user_id}'. Please try another user."
    
    # Get the user's ratings
    user_ratings = pred_df[pred_df['uid'] == user_id]

    # Sort the ratings by score and get the top n
    top_n_products = user_ratings.nlargest(n, 'est')['iid'].tolist()
    
    return top_n_products, None

def hybrid_recommend(user_id, product_name, alpha=0.5, n=10):
    # Get the collaborative filtering recommendations
    collaborative_products, _ = get_collaborative_score(user_id, pred_df, n=n)
    
    # Get the content-based recommendations using product name
    content_products, _ = recommend_products(product_name, cosine_sim_df, product_info_copy, n=n)
    
    # Combine the two recommendation sets using the hybrid method
    recommended_products = {}
    
    for i, product in enumerate(collaborative_products):
        if product in recommended_products:
            recommended_products[product] += alpha * (n - i)
        else:
            recommended_products[product] = alpha * (n - i)
    
    for i, product in enumerate(content_products):
        if product in recommended_products:
            recommended_products[product] += (1 - alpha) * (n - i)
        else:
            recommended_products[product] = (1 - alpha) * (n - i)
    
    # Sort the combined recommendations by the hybrid score
    sorted_products = sorted(recommended_products.items(), key=lambda x: x[1], reverse=True)
    
    # Return the top N products and their corresponding hybrid scores
    return sorted_products[:n]

```
